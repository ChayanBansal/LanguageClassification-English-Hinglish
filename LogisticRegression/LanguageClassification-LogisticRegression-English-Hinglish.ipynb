```python
import pandas as pd

df_en = pd.read_csv('..\data\english_text.csv')
df_hi = pd.read_csv('..\data\hinglish_text.csv')
df_en['label'] = 'english'
df_hi['label'] = 'hinglish'
df = pd.concat([df_en,df_hi])
```

```python
df_tagged_data = pd.read_csv('..\data\hinglish_english_text.csv', header=None, names = ['ID','text','label_id'])
df_tagged_data.loc[df_tagged_data['label_id']==1, 'label'] = 'hinglish'
df_tagged_data.loc[df_tagged_data['label_id']==0, 'label'] = 'english'
df_tagged_data.head()
```

```python
df = pd.concat([df,df_tagged_data[['ID','text','label']]])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(min_df=0, lowercase=False)
vectorizer.fit(df['text'])
vectorizer.vocabulary_
```

```python
from sklearn.model_selection import train_test_split


sentences = df['text'].values
y = df['label'].values

sentences_train, sentences_test, y_train, y_test = train_test_split(sentences, y, test_size=0.8, random_state=1000)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
vectorizer.fit(sentences_train)

X_train = vectorizer.transform(sentences_train)
X_test  = vectorizer.transform(sentences_test)
X_train
```

```python
from sklearn.linear_model import LogisticRegression

classifier = LogisticRegression()
classifier.fit(X_train, y_train)
score = classifier.score(X_test, y_test)

print("Accuracy:", score)
```
